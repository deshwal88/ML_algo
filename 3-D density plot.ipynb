```python
from mpl_toolkits.mplot3d import Axes3D

def limits(x):
    w=max(x)-min(x)
    up=max(x)+0.3*w
    low=min(x)-0.3*w
    return low,up

def measure_bi(x,feature):
    mean=np.mean(x[:,feature],axis=0)
    cov=np.cov(x[:,feature],rowvar=False,bias=True)
    return mean,cov
    
```

```python
def plot(x,feature):
    x1_lim=limits(x[:,feature[0]])
    x2_lim=limits(x[:,feature[1]])
    x1=np.linspace(x1_lim[0],x1_lim[1],200)
    x2=np.linspace(x2_lim[0],x2_lim[1],200)
    X1,X2=np.meshgrid(x1,x2)
    z=np.zeros(X1.shape)
    mean,cov=measure_bi(x,feature)
    multi=multivariate_normal(mean=mean,cov=cov)
    
    for i in range(0,len(X1)):
        for j in range(0,len(X2)):

            z[j,i]=multi.logpdf([x1[i],x2[j]])
    

    fig=plt.figure()
    ax = fig.gca(projection='3d')
    ax.set_xlim(x1_lim[0],x1_lim[1])
    ax.set_ylim(x2_lim[0],x2_lim[1])
    ax.plot_surface(X1,X2,z)
    ax.view_init(azim=45, elev=40)
```

```python
plt.rcParams['figure.figsize']=(12,9)
plot(trainx[trainy==1,:],[0,6])
```
