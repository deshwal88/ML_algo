```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
header=['sepal-length','sepal-width','petal-length','petal-width','category']
data=pd.read_csv('/home/kanishk/Downloads/iris.csv',names=header)
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=data.drop(labels=['category'],axis=1)
y=data['category']
sc.fit(x)
x=sc.transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.75,random_state=3)
```

```python
from sklearn.metrics import accuracy_score as score, confusion_matrix as cm, classification_report as cr
```

```python
from sklearn.tree import DecisionTreeClassifier
dsc=DecisionTreeClassifier(criterion='entropy')
```

```python
dsc.fit(x_train,y_train)
y_hat=dsc.predict(x_test)
print(cr(y_hat,y_test))
```

```python
from sklearn import tree
tree.plot_tree(dsc)
plt.figure(figsize=(12,6))
plt.show()
from sklearn.preprocessing import LabelEncoder
```
