```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
url='https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
database=pd.read_csv(url,names=['sepal length','sepal width','petal length','petal width','class'])
database.head()
print(database.shape)
X=database.iloc[:,1].values
Y=database.iloc[:,2].values
X_mean=np.mean(X)
Y_mean=np.mean(Y)


num,denom=0,0
for i in range(len(X)):
    num+=(X[i]-X_mean)*(Y[i]-Y_mean)
    denom+=(X[i]-X_mean)**2

m=num/denom
c=Y_mean-(m*X_mean)

x=np.linspace(0,(np.max(X)),100)
y=m*x+c

plt.plot(x,y,label='Regression Line')
plt.scatter(X,Y,label='Scatter plot')
plt.xlabel('Sepal Length')
plt.ylabel('sepal width')
plt.legend()
plt.show()

var_pre,var_tot=0,0
for i in range(len(Y)):
    var_pre+=(Y[i]-(m*X[i]+c))*2
    var_tot+=(Y[i]-Y_mean)*2
R_square=1-(var_pre/var_tot)
print(R_square)
```
