## Loss functions: Entropy and Cross entropy 

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
from mpl_toolkits.mplot3d import axes3d
warnings.filterwarnings('ignore')
%matplotlib inline
```

### Defining entropy functions

```python
def entropy(arr):
    l0=0
    for i in arr:
        lt=-i*np.log(i)
        if np.isnan(lt) or np.isinf(lt):
            yield l0
        else:
            l0=lt
            yield l0
            
def cross_entropy(arr1,arr2):
    l0=0
    for i,j in zip(arr1,arr2):
        lt=-i*np.log(j)
        if np.isnan(lt) or np.isinf(lt):
            yield l0
        else:
            l0=lt
            yield l0
    
```

```python
n=np.array([i for i in range(1,101)])/100
_iter_=entropy(n)
ent=[_iter_.__next__() for i in range(0,len(n))]
```

### Visualizing Entropy

```python
plt.figure(figsize=(6,4))
plt.plot(n,ent)
plt.xlabel('Probablity (Pi)')
plt.ylabel('Entropy')
plt.title('Distribution of entropy over different probablities')
plt.show()
```

As one can see when the probablity of an event takes extreme values, entropy is minimum as it is certain that the event will occur or not. But in the middle, say 0.5 probablity, it is uncertain thus the entropy is maximum.

```python
cross_ent=np.zeros((len(n),len(n)))
for i in range(0,cross_ent.shape[0]):
    _iter_2=cross_entropy(np.ones(len(n))*n[i],n)
    cross_ent[i]=[_iter_2.__next__() for i in range(0,len(n))]
cross_ent.shape
```

### Visualizing cross entropy

```python
x,y=np.meshgrid(n,n)
fig=plt.figure(figsize=(8,6))
ax=axes3d.Axes3D(fig)
ax.plot_surface(x,y,cross_ent)
plt.xlabel('Probablity (P2)')
plt.ylabel('Probablity (P1)')
ax.set_zlabel('Cross-entropy')
plt.title('Distribution of cross-entropy over different probablities')
plt.show()
```

The 3D figure here shows the effect of cross entropy and probablity values P1 and P2.
