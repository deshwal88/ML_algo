```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('ggplot')
sns.set_style('white')
plt.rcParams['figure.figsize']=(5,5)
plt.figure()

url='https://raw.githubusercontent.com/JangirSumit/data_science/master/Hierachical%20Clustering/movie_metadata1.csv'
data=pd.read_csv(url)
new_data=data[['director_facebook_likes','actor_3_facebook_likes']]


from sklearn.cluster import KMeans
model=KMeans(n_clusters=4)
model.fit(new_data)
l=model.labels_
c=model.cluster_centers_
print(l,'\n',c)
clusters,counts=np.unique(l,return_counts=True)
d=dict(zip(clusters,counts))
print(d)
new_data.insert(2,'clusters',l)
sns.lmplot(x='director_facebook_likes',y='actor_3_facebook_likes',hue='clusters',fit_reg=False,data=new_data,palette='coolwarm')
plt.show()
```
